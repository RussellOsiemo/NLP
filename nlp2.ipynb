```python
from bs4 import BeautifulSoup
import os
import re
# text1 = open('amazon.txt', 'r')
# text2 = open('global.hitachi-solutions.txt', 'r')
# text3 = open('googlecloud.txt', 'r')
# text4 = open('forseemed.txt', 'r')
# text5 = open('healthcatalyst.txt', 'r')
# text6 = open('lexalytics.txt', 'r')
# Cleaning up Data Scraped from the Web
# Cleaning HTML 
# Function to remove tags                
# Strip whitespace
def clean_html_tags(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()
         # Remove HTML, CSS, and JavaScript code
        cleaned_content = re.sub(r'<[^>]*>|{[^}]*}|\[[^\]]*\]', '', content)
          # Remove non-English letters
        cleaned_content = re.sub(r'[^a-zA-Z]', ' ', cleaned_content)
        # remove whitespace
        cleaned_content = cleaned_content.strip()
    return cleaned_content
def clean_html_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            file_path = os.path.join(directory, filename)
            cleaned_content = clean_html_tags(file_path)
            
            new_file_path = os.path.join(directory, 'cleaned_' + filename)
            with open(new_file_path, 'w', encoding='utf-8') as f:
                f.write(cleaned_content)
                
directory_path = '/content'
clean_html_files(directory_path)



# Converting numbers to number types:
# Converting Boolean values: ‘Yes’ -> True
# Converting dates to machine-readable formats: “24 June 2004” -> “2004-06-24”
```

```python
# tokenization
import nltk
nltk.download('punkt')
# nltk.download()
dir_path = '/content'

# Loop over each file in the directory
for file_name in os.listdir(dir_path):
    if file_name.endswith('.txt') and file_name.startswith("cleaned"):  # Check if the file is a text file
        file_path = os.path.join(dir_path, file_name)
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()
            tokens = nltk.word_tokenize(text)
            # Do whatever you want with the tokens, such as counting them
            print(f'File named: {file_name} has {len(tokens)} tokens.')
```

```python
# turn the text to lowercase
import os

def lowercase_cleaned_files(dir_path):
    # Loop over each file in the directory
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.txt') and file_name.startswith('cleaned'):  # Check if the file is a text file and starts with "cleaned"
            # Define input and output file paths
            input_file_path = os.path.join(dir_path, file_name)
            output_file_name = f"lowercased_{file_name}"
            output_file_path = os.path.join(dir_path, output_file_name)
            
            # Read the input file and convert to lowercase
            with open(input_file_path, 'r') as f:
                text = f.read()
                lower_text = text.lower()
            
            # Write the lowercase text to a new file
            with open(output_file_path, 'w') as f:
                f.write(lower_text)
# call the function
dir_path = '/content'
lowercase_cleaned_files(dir_path)
```

```python
# removing stop words
import os
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords_from_files(dir_path):
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.txt') and file_name.startswith('lowercased'):
            input_file_path = os.path.join(dir_path, file_name)
            output_file_name = f"stopwords_removed_{file_name}"
            output_file_path = os.path.join(dir_path, output_file_name)
            
            with open(input_file_path, 'r') as f:
                text = f.read()
                tokens = word_tokenize(text)
                filtered_tokens = [word for word in tokens if not word in stop_words]
                filtered_text = ' '.join(filtered_tokens)
                
            with open(output_file_path, 'w') as f:
                f.write(filtered_text)

dir_path = '/content'
remove_stopwords_from_files(dir_path)
```

```python
# plotting the distribution
import os
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
import seaborn as sns
def plot_word_frequency(dir_path):
    # Initialize empty list to store all tokens
    all_tokens = []
    
    # Loop over each file in the directory
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.txt') and file_name.startswith('stopwords'):
            # Define input file path
            input_file_path = os.path.join(dir_path, file_name)
            
            # Read the input file and tokenize the text
            with open(input_file_path, 'r') as f:
                text = f.read()
                tokens = word_tokenize(text)
                
            # Add the tokens to the list of all tokens
            all_tokens.extend(tokens)
    
    # Compute frequency distribution of all tokens
    fdist = FreqDist(all_tokens)
    
    # Plot a histogram of the top 20 most frequent words
    plt.figure(figsize=(10, 6))
    sns.set_style('darkgrid')
    fdist.plot(20)
    
    plt.show()
dir_path = '/content'
plot_word_frequency(dir_path)
```

```python
import os
from nltk.tokenize import word_tokenize

def count_unique_words(dir_path):
    # Initialize empty set to store unique words
    unique_words = set()
    
    # Loop over each file in the directory
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.txt') and file_name.startswith('stopwords_removed_'):
            # Define input file path
            input_file_path = os.path.join(dir_path, file_name)
            
            # Read the input file and tokenize the text
            with open(input_file_path, 'r') as f:
                text = f.read()
                tokens = word_tokenize(text)
                
            # Add the unique tokens to the set of unique words
            unique_words.update(set(tokens))
    
    # Return the number of unique words
    return len(unique_words)

dir_path = '/content'
num_unique_words = count_unique_words(dir_path)
print(f'The number of unique words is: {num_unique_words}')
```

```python
import os
import pandas as pd
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer 
def tf_idf_calculator(dir_path):
     # Initialize empty lists to store document names and document text
    doc_names = []
    doc_text = []
    
    # Loop over each file in the directory
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.txt') and file_name.startswith('stopwords_removed_'):
            # Define input file path
            input_file_path = os.path.join(dir_path, file_name)
            
            # Read the input file and tokenize the text
            with open(input_file_path, 'r') as f:
                text = f.read()
                tokens = word_tokenize(text)
                doc_text.append(' '.join(tokens))
            
            # Add the document name to the list of document names
            doc_names.append(file_name)
    
    # Compute tf-idf scores for the documents
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(doc_text)
    
    # Store the tf-idf scores in a Pandas DataFrame
    tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
    tfidf_df.index = doc_names
    
    return tfidf_df

dir_path = '/content'
tf_idf = tf_idf_calculator(dir_path)
print(tf_idf)
```

```python
import os
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer

def compute_tf_idf(dir_path):
    # Initialize empty dictionary to store top 10 words with highest tf-idf values for each document
    top_words = {}
    
    # Loop over each file in the directory
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.txt') and file_name.startswith('stopwords_removed_'):
            # Define input file path
            input_file_path = os.path.join(dir_path, file_name)
            
            # Read the input file and tokenize the text
            with open(input_file_path, 'r') as f:
                text = f.read()
                tokens = word_tokenize(text)
                text = ' '.join(tokens)
                
            # Compute tf-idf scores for the document
            tfidf_vectorizer = TfidfVectorizer()
            tfidf_matrix = tfidf_vectorizer.fit_transform([text])
            
            # Get the top 10 words with the highest tf-idf scores
            feature_names = tfidf_vectorizer.get_feature_names()
            sorted_tfidf_scores = sorted(enumerate(tfidf_matrix.toarray()[0]), key=lambda x: x[1], reverse=True)
            top_words[file_name] = [feature_names[idx] for idx, _ in sorted_tfidf_scores[:10]]
    
    return top_words
dir_path = '/content'
top_words = compute_tf_idf(dir_path)

# Print the top 10 words with the highest tf-idf values for each document
for doc, words in top_words.items():
    print(doc)
    print(words)
    print('\n')
```

```python
import os
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def compute_cosine_similarity(dir_path):
    # Initialize empty list to store document texts
    documents = []
    document_names = []
    
    # Loop over each file in the directory
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.txt') and file_name.startswith('stopwords_removed_'):
            # Define input file path
            input_file_path = os.path.join(dir_path, file_name)
            
            # Read the input file and tokenize the text
            with open(input_file_path, 'r') as f:
                text = f.read()
                tokens = word_tokenize(text)
                text = ' '.join(tokens)
                documents.append(text)
                document_names.append(file_name)
                
    # Compute tf-idf scores for all the documents
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(documents)
    
    # Compute cosine similarity matrix
    similarity_matrix = cosine_similarity(tfidf_matrix)
    
    # Find the top 5 most similar documents for each document
    num_documents = len(documents)
    top_similar_documents = {}
    for i in range(num_documents):
        document_similarities = list(enumerate(similarity_matrix[i]))
        sorted_similarities = sorted(document_similarities, key=lambda x: x[1], reverse=True)
        top_similar_documents[document_names[i]] = [document_names[j] for j, _ in sorted_similarities[1:6]]
    
    return top_similar_documents
dir_path = '/content'
top_similar_documents = compute_cosine_similarity(dir_path)

# Print the top 5 most similar documents for each document
for i, similar_docs in top_similar_documents.items():
    print(f"Most similar documents to document {i}:")
    for doc_idx in similar_docs:
        print(f"  {doc_idx}")
    print('\n')
```
